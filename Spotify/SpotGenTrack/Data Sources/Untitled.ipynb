```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv("spotify_artists.csv")
display(dados)
```

```python
dados.drop('id',axis=1, inplace=True)
dados.drop('Unnamed: 0',axis=1, inplace=True)
dados.drop('track_id',axis=1, inplace=True)
dados.drop('track_name_prev',axis=1, inplace=True)
dados.drop('type',axis=1, inplace=True)
display(dados)
```

```python
select= dados.loc[dados['artist_popularity'].values >60]
populares=select.sort_values('artist_popularity',ascending=False)
populares50 = populares.head(50)
populares50
```

```python
plt.figure(figsize=(20,4))
plt.bar(populares50['name'],populares50['artist_popularity'],color='r')
plt.xticks(populares50['name'],rotation=90)
plt.title("Os 50 artistas mais polpulares no Spotify",fontsize=50)
plt.xlabel("Artistas",fontsize=40)
plt.ylabel("Popularidade",fontsize=40)

plt.savefig("artistas populares.png")

plt.show()
```
